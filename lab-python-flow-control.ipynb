```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}
for product in products: 
    while True:
        try:
            quantity=int(input(f"Please enter the number of units available of {product}: "))

            if quantity < 0: 
                print(f"Please enter a valid whole number of units of {product} available ")
                continue
            inventory[product]= quantity
            break

        except ValueError:
            print(f"Invalid input. Please enter a valid whole number of units of {product} available ")
            
print("The original inventory is: ",inventory)
customer_orders = set()
shopping = True

while shopping:
    order_1 = input("Write the name of the product you would like to order: ").strip().lower()
    if order_1 not in products:
                print("Item name couldnt be found. Please try again:")
                continue
    else:
        customer_orders.add(order_1)
        print(f"{order_1} has been added to the basket." )
        while True:
            print("Would you like to keep purchasing on our web?")
            purchase = input("Press '1' for Yes '2' for No: ")
            if purchase == "1":
                break
            elif purchase == "2":
                shopping = False
                break
            else:
                 print("Invalid option.")
        
    
print(f"Your purchase list is: {customer_orders}")
total_products_ordered = len(customer_orders)

percentage_of_total_products = total_products_ordered/len(products)*100

order_status = (total_products_ordered,percentage_of_total_products)

print("Order Statistics: ")

print(f"The total number of products ordered is: ",total_products_ordered)

print(f"The percentage of the total of this order is: {percentage_of_total_products} %")

for product in customer_orders:
    inventory[product] -= 1
print("Updated inventory: ")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
```
